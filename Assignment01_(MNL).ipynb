**Question 1.1**
Write a Python function that takes two positive integers and returns their greatest common divisor. In case you pass a negative integer to the function it must return the following string in red: "This function takes only positive integers!

````python
from collections import Counter

def ALG_Euclide(positive_integer_1, positive_integer_2):
  # Returns the greatest common divisor of two positive integers

  if positive_integer_1 < 1 or positive_integer_2 < 1:                                                # Conditional: If pass negative integer to the function
    return print(bcolors.RED + "This function takes only positive integers!" + bcolors.ENDC)          # Fail message for the function
  
  else:
    x = []                                                                                            # Initialize list for x and y
    y = []

    for i in range(1, positive_integer_1 + 1):                                                        # Loop for i in integer 1 to variable_1 + 1 in function
      if positive_integer_1 % i == 0:                                                                 # Conditional: if variable_1 in function / i = integer (modulo);
        x.append(i)                                                                                   # Add this number in the list vector for x

    for i in range(1, positive_integer_2 + 1):                                                        # Same loop, but for the variable_2 in the function
      if positive_integer_2 % i == 0:
        y.append(i)

    xy = x+y                                                                                          # Create vector xy container all numbers in vector x and y
    counts = Counter(xy)                                                                              # Count the number of integer in the vector xy 

  return max([x_i for x_i, count in counts.items() if count == 2])                                    # Comprehension List: For all x_1 in items of counts, return maximum numbre if count == 2
````

**Question 1.2** Write a function that prints all the prime numbers less than or equal to  n , where  n  is a parameter to be passed to the function. If  nâ‰¤1  the function returns "No prime number was found!".

````python
def prime_nbr(nombre):
  # Prints all the prime numbers less than or equal to n

  if nombre <= 1:                                                                    # Conditional: If n <= 1, fail message
    return print(bcolors.RED + "No prime number was found!" + bcolors.ENDC)          # Fail message for the function

  for n in range(2, nombre + 1):                                                     # Loop for n in integer 2 to variable + 1 of the function
    for i in range(2,n):                                                             # Loop for i in integer 2 to n
      if n % i == 0:                                                                 # Conditional: if n / i = integer (modulo)
        break                                                                        # Terminates the current loop and resumes execution at the next statement
    else:
      print(n)                                                                       # Print the prime number n
````

In probability theory, the law of large numbers (LLN) is a theorem that describes the result of performing the same experiment a large number of times. According to the law, the average of the results obtained from a large number of trials should be close to the expected value and will tend to become closer to the expected value as more trials are performed. 

The more we increase the number of observations in the sample, the closer that the estimated population mean should be to the population mean (0.83333333333).

**Question 3.5** Fit a least squares linear model to predict y using x . Comment on the model obtained. How do  Î²^0  and  Î²^1  compare to  Î²0  and  Î²1 ?

````python
# Estimate the intercept and the regressor's coefficient with the least square solution.
x_bar = sum(x)/len(x)
y_bar = sum(y)/len(y)

a_num = sum([(x_i - x_bar)*(y_i - y_bar) for x_i, y_i in zip(x,y)])
a_denom = sum([(x_i - x_bar)**2 for x_i in x])

reg_coef = a_num/a_denom
intercept = y_bar - reg_coef*x_bar

print(reg_coef)
print(intercept)
````

For the population:
- Intercept (b0) = -1
- Regressor's coefficient (b1) = 0.5

For the sample:
- Intercept (b^0) = -0.9776
- Regressor's coefficient (b^1) = 0.4731

Difference with the population and the sample is occasionate by the noise (eps ~ N(0, 0.25)), but the intercept b^0 and the slope b^1 of the sample are near to the population.

**Question 3.7** Explain how the result would change if you generate the data with less noise and repeat all steps 1 to 6? What about more noise?

````python
np.random.seed(113)
mu, sigma = 0, 0.25 #population
n = 100
eps_2 = np.random.normal(mu, sigma, n) # sample with noise less variance

y = -1 + (0.5*x) + eps_2

# Estimate the intercept and the regressor's coefficient with the least square solution.
x_bar = sum(x)/len(x)
y_bar = sum(y)/len(y)

a_num = sum([(x_i - x_bar)*(y_i - y_bar) for x_i, y_i in zip(x,y)])
a_denom = sum([(x_i - x_bar)**2 for x_i in x])

reg_coef = a_num/a_denom
intercept = y_bar - reg_coef*x_bar

print(reg_coef)
print(intercept)

# Draw the scatterplot and the regression equation together.

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(x , y, 'or', mfc='b')
line, = ax.plot(np.array(x),intercept +  reg_coef * np.array(x), color='orange', label='Sample')
line, = ax.plot(np.array(x),-1 +  0.5 * np.array(x), color='g', label='Population')
ax.set_title('Linear Regression: Population and Sample', fontsize=14)
ax.set_ylabel('Variable y', fontsize=12)
ax.set_xlabel('Variable x', fontsize=12)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
plt.grid(True)
````

With less noise, the response variable for the sample has less variance. If we add most noise, it's the contrary effet: the response variable for the sample has more variation.


The biasâ€“variance tradeoff is the property of a model that the variance of the parameter estimates across samples can be reduced by increasing the bias in the estimated parameters.


The bias error is an error from erroneous assumptions in the learning algorithm. High bias can cause an algorithm to miss the relevant relations between features and target outputs (underfitting).


The variance is an error from sensitivity to small fluctuations in the training set. High variance can cause an algorithm to model the random noise in the training data, rather than the intended outputs (overfitting).
